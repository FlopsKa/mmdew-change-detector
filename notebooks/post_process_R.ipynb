```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Cov shift

```python
statistics = []

for i in range(1,21):
    statistics += [list(pd.read_csv(f"results/R/cov_shift/CvM/out{i}.csv", sep=" ").mean())]
```

```python
stats = np.max(np.array(statistics),axis=0)
```

```python
plt.plot(stats)
```

```python
cov = pd.DataFrame(stats, columns=["test_stat"])
cov["algorithm"] = "CPM"
cov["experiment"] = "Covariance shift"
```

# Mean shift

```python
statistics = []

for i in range(1,2):
    statistics += [list(pd.read_csv(f"results/R/mean_shift/CvM/out{i}.csv", sep=" ").mean())]

stats = np.max(np.array(statistics),axis=0)
```

```python
plt.plot(stats)
```

```python
mean = pd.DataFrame(stats, columns=["test_stat"])
mean["algorithm"] = "CPM"
mean["experiment"] = "Mean shift"
```

```python
df = pd.concat((cov,mean))

df.to_csv("results/cpm-univariate.csv")
```
